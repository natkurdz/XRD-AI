LINE 1 : 

    TITLE  (any 70 characters to be used to label the printout) 

LINE 2 : 

    !Job Npr Nph Nba Nex Nsc Nor Dum Iwg Ilo Ias Res Ste Nre Cry Uni Cor 

    (15 integers - It is understood that they are separated by a space) 

JOB    = 		0   X-ray case 

1   Neutron case (constant wavelength, nuclear and magnetic) 

        		2   pattern calculation (X-ray) 

        		3   pattern calculation (Neutron, constant wavelength) 

       		-1   Neutron case (T.O.F., nuclear and magnetic) 

       		-3   pattern calculation (Neutron, T.O.F.) 

 

NPR  = 	 0  Gaussian 

        	1  Cauchy 

        	2  Modified 1 Lorentzian 

        	3  Modified 2 Lorentzian  

        	4  Tripled pseudo-Voigt 

        	5  pseudo-Voigt 

        	6  Pearson VII  

        	7  Thompson-Cox-Hastings pseudo-Voigt – most suitable for X-ray diffraction 

        	8  Numerical profile given in CODFIL.SHP 

           	or in GLOBAL.SHP 

        	9  T.O.F. Convolution pseudo-Voigt x Double Exponential 

       	10  Not yet used 

       	11  Split pseudo-Voigt function 

       	12  Pseudo-Voigt function convoluted with axial divergence asymmetry 

           	function (Finger, Cox & Jephcoat, J. Appl. Cryst. 27, 892, 1994) 

 

NPH =   number of phases ( max:8)  

NBA  =  0  Refine background with polynomial function 

  	1        Read background from file CODFIL.BAC.  

        	2,3,.,N  linear interpolation between the N given points 

 

NEX     =   number of excluded regions 

IWG     =	0 standard least squares refinement 

1 maximum likelihood refinement 

      		2 unit weights 

 

 

 

NRE      Number of parameters to be constrained within given limits. At the end of the file you must give a list of NRELL lines specifying the number and the limit of each parameter. 

 

LINE 3 : 

!Ipr Ppl Ioc Mat Pcr Ls1 Ls2 Ls3 Syo Prf Ins Rpa Sym Hkl Fou Sho Ana    (17 integers) 

List of output control flags : normally 0 = off / any value = on 

 

IPR   = 1 obs. & calc. profile intensities -->  CODFIL.OUT (0) 

            2 The files CODFILn.SUB with the calculated profile of 

               each phase are generated. 

            3 As 2 but the background is added to each profile. 

 

PPL   = 1 line printer plot  --> CODFIL.OUT 

             2 Generates the background-file FILE.BAC 

             3 Puts difference pattern in file FILE.BAC 

 

IOC   = 1 list of obs. & calc. integr. int.  --> CODFIL.OUT 

         = 2 The reflections corresponding to the second wavelength 

          are also written if different from the first one. 

 

PCR  = 1 CODFIL.PCR is re-written with updated parameters 

            2 new input file  --> CODFIL.NEW 

 

INS   =   9   Siemens and Bruker UXD format 

      	10  X,Y,Sigma format with header lines. 

              	13 PANAlytical format xrdml 

 

LINE 4  

! lambda1 Lambda2    Ratio   BKPOS    Wdt    Cthm     muR   AsyLim   Rpolarz        (9 reals)    

 

LAMBDA1   = wavelength l1 

 

LAMBDA2   = wavelength l2 (= l1 for monochromatic beam) 

 

RATIO = Intensity ratio I(l2)/I(l1) for typical Cu Kalpha1,2 this is equal to 0.5 

           If RATIO<0 the parameters U, V, W (see below) for the 

           second wavelength are read separately. 

 

LINE 5  

!NCY  Eps R_at  R_an  R_pr  R_gl     Thmin    Step    Thmax    PSD     

    (1 integer and 10 reals) 

 

NCY   =   number of cycles of refinement 

EPS =      forced termination when shifts less than EPS*e.s.d. 

 

 

LINE 10.1  

!  Zero    Code    Sycos    Code   Sysin    Code  Lambda     Code MORE 

 

  ZERO     = zero point for T (in degrees) : Ttrue = Texp.-ZER, zero shift of the instrument 

  CODE  = codeword for zeroshift  

  SYCOS   = systematic 2Theta shift with cosTheta dependenc sample displacement (Theta - 2Theta diffractometers) 

  CODE   = codeword for SYCOS 

  SYSIN   = systematic 2Theta shift with sin2Theta dependence of sample transparency coefficient 

  CODE   = codeword for SYSIN 

 

LINE 10.2 (only if Nba = 0) 

!   Background coefficients/codes  

 

Firts row – six parameters of the polynomial A*x+B*x^1+C*x^2+...+F*x^5 

Second row codes for each A, B, ...F parameters  

 

LINE 11.1 

Name of the subsequent phase 

 

LINE 11.2 

!Nat Dis Mom Pr1 Pr2 Pr3 Jbt Irf Isy Str Furth     ATZ   Nvk Npr More 

NAT = number of atoms within the unit cell 

JBT        = 0 The phase is treated with the Rietveld Method, then 

                      refining a given structural model. 

              = 2 Profile Matching mode with constant scale factor (only space group and lattice parameters will be refined) 

 

IRF        = 0 The list of reflections for this phase is automatically generated from the space group   

           = 1 The list h,k,l,Mult is read from file CODFILn.HKL (where n 

               is the ordinal number of the current phase) 

ISYM = 0    The symmetry operators are generated automatically from the space group symbol. 

NPRO = Integer indicating the peak shape for the present phase (see line 2). If NPRO=0, the default value NPROF is taken. 

 

LINE 11.3  

Symbol of space group for current phase 

 

LINE 11.4 

!Atom    Typ       X        Y        Z     Biso      Occ     In Fin N_t /Codes 

ATOM – name of the atom (eg. Fe1) 

TYP – usually a name from periodic table (capital letters) – eg. FE, IN, PD 

X, Y, Z – fractional coordinates 

BISO – isotropic temperature factor (Deby’e Waller factor) 

OCC – occupation of the site = real occupation (in between 0 and 1) * site mutiplicity / general multiplicity (the highest possible within given space group – for position x,y,z) 

Second row – respective codes 

 

LINE 11.5.1 

!  Scale       Shape1      Bov     Str1     Str2     Str3    Strain-Model 

  SCALE               = scale factor 

  SHAPE1            = profile shape parameter , e.g. : 

                    eta0 for NPROF=4,5 but not for  NPROF=7 (see line 11-6-1), for NPROF = 7, SHAPE =0 

                    m0 for NPROF=6 

  BOV            = overall isotropic displacement (temperature) factor in angstroms**2 

LINE 11.6.1 

!      U      V       W        X       Y     GauSiz   LorSiz Size-Model 

Shape of the relations related parameters. 

U, V, W – instrumental broadening 

X, Y – sample broadening, in simple model: X –strains, Y – crystallite size. 

For NPROF = 7, the FWHM of the two components is calculated as 

FWHM^2(gaussian) = (U)*Tan^2(Theta) + V*Tan(Theta) + W  

FWHM(lorentzian) =  X tan(Theta) + (Y)/cos(Theta) 

 

 

, where K – is Scherrer shape factor (you can assume K = 1), [Równanie] – is the wavelength, it must be given in angstroms in order to get SIZE in angstroms as well. 

 

 

LINE 11.7.1 

!     a          b         c        alpha      beta       gamma 

Lattice parameters (in angstroms) and angles 

Second row codes – for gamma= 120 (hexagonal) the codes for a and b must be the same as for gamma (of course gamma will not be refined)
